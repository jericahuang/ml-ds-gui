## 2. Get familiar with real data

Astronomers describe a star's position in the sky by using a pair of angles: declination and right ascension. Declination is similar to longitude, but it is projected on the celestian fear. Right ascension is known as the "hour angle" because it accounts for time of day and earth's rotaiton. Both angles are relative to the celestial equator. You can learn more about star position [here](https://en.wikipedia.org/wiki/Star_position).

````python
#Scorpius Constellation
#JSON string to pair every star in the constellation with its right ascension, declination, and distance
#star information retrieved from Wikipedia
scorpius = {
    # star name: {
    #    right ascension: (hours, minutes, seconds),
    #    declination: (degress, minutes, seconds),
    #    distance : parsecs 
    #    },
    "Beta Scorpii":    { "asc": (16, 5, 26.23198),    "dec": (-19, 48, 06.8556),    "dis": 120 }, # 1st Horn
    "Delta Scorpii":   { "asc": (16, 0, 20.00528),    "dec": (-22, 37, 18.1431),    "dis": 136 }, # 2nd Horn
    "Pi Scorpii":      { "asc": (15, 59, 51.11324),   "dec": (-26, 6, 50.7886),     "dis": 180 }, # 3rd Horn
    "Sigma Scorpii":   { "asc": (16, 21, 11.31571),   "dec": (-25, 35, 34.0515),    "dis": 174 },
    "Alpha Scorpii":   { "asc": (16, 29, 24.45970),   "dec": (-26, 25, 55.2094),    "dis": 170 }, # Antares
    "Tau Scorpii":     { "asc": (16, 35, 52.95285),   "dec": (-28, 12, 57.6615),    "dis": 150 }, # The vagabond
    "Epison Scorpii":  { "asc": (16, 50, 9.8),        "dec": (-34, 17, 36),         "dis": 20 },
    "Mu1 Scorpii":     { "asc": (16, 41, 52.23111),   "dec": (-38, 2, 50.5694),     "dis": 150 },
    "Mu2 Scorpii":     { "asc": (16, 52, 20.14532),   "dec": (-38, 1, 3.1258),      "dis": 145 },
    "Zeta1 Scorpii":   { "asc": (16, 53, 59.72650),   "dec": (-42, 21, 43.3063),    "dis": 800 },
    "Zeta2 Scorpii":   { "asc": (16, 54, 35.00435),   "dec": (-42, 21, 40.7407),    "dis": 41 },
    "Eta Scorpii":     { "asc": (17, 12, 9.19565),    "dec": (-43, 14, 21.0905),    "dis": 23 },
    "Theta Scorpii":   { "asc": (17, 37, 19.12985),   "dec": (-45, 59, 52.1808),    "dis": 90 },
    "Iota1 Scorpii":   { "asc": (17, 47, 35.08113),   "dec": (-40, 7, 37.1893),     "dis": 590 },
    "Iota2 Scorpii":   { "asc": (17, 50, 11.11291),   "dec": (-40, 5, 25.5629),     "dis": 800 },
    "Kappa Scorpii":   { "asc": (17, 42, 29.27520),   "dec": (-39, 1, 47.9391),     "dis": 148 },
    "Upsilon Scorpii": { "asc": (17, 30, 45.83712),   "dec": (-37, 17, 44.9285),    "dis": 177 }, # The tail    
    "Lambda Scorpii":  { "asc": (17, 33, 36.520),     "dec": (-37, 6, 13.76),       "dis": 180 }, # The sting
}
````

````python
#Convert a star's right ascension, declination, and distance into its x,y,z co-ordinates
from math import sin, cos, radians

# Convert Right Ascension, Declination and Distance into x, y, z co-ordinates
# params:
#     asc - right ascension passed as a tuple of (hours, minutes, seconds)
#     dec - declination passed as a tuple of (degrees, minutes, seconds)
#     dis - distance
def radd_to_xyz(asc, dec, dis):
    # convert right ascension to radians - cos & sin take radians as params
    def asc_to_rad(hours, minutes, seconds):
        return radians((hours*15) + (minutes/4) + (seconds/240))
    # convert declination to radians
    def dec_to_rad(degrees, minutes, seconds):
        sign = 1 if degrees > 0 else -1
        return radians((abs(degrees) + (minutes/60) + (seconds/3600)) * sign)
    # Pass tuples as params by preceeding with '*'
    asc_r = asc_to_rad(*asc)
    dec_r = dec_to_rad(*dec)
    rvect = dis * cos(dec_r)
    x = rvect * cos(asc_r)
    y = rvect * sin(asc_r)
    z = dis * sin(dec_r)
    return (x, y, z)

# Test this with the example given on the Agorithm page.
# Tau Ceti is at Right Asc 1h41m45s, Dec -16Â°12m, Distance 3.61 parsecs - worked example gives X= 3.13, Y=1.49, Z=-1.01
tau_ceti = {"asc": (1, 41, 45), "dec": (-16, 12, 0), "dis": 3.61 }

# Pass dictionary as parameter by preceeding with '**'
x, y, z = radd_to_xyz(**tau_ceti)

print("Expected x = 3.13, y = 1.49, z = -1.01")
print("Actual   x = {0:2.2f}, y = {1:2.2f}, z = {2:2.2f}".format(x, y, z))
````

````python
#combine these together to get x, y, z co-ordinates
x_values = []
y_values = []
z_values = []
for star in scorpius.keys():
    x, y, z = radd_to_xyz(**scorpius[star])
    
    # for scorpius all values are negative, which upsets the 3D plot, so make them positive
    x_values.append(abs(x/10))
    y_values.append(abs(y/10))
    z_values.append(abs(z/10))
    
for x,y,z in list(zip(x_values, y_values, z_values)):
    print("{0:03.4f}, {1:03.4f}, {2:03.4f}".format(x, y, z))
````

````python
fig = plt.figure()
fig.add_subplot(1, 1, 1)
plt.scatter(x_values, y_values, color="blue")
plt.title("2D image of Scorpius (X and Y Coordinates)")
plt.xlabel("X coordinates")
plt.ylabel("Y coordinates")
plt.show()
````

````python
fig_3d = plt.figure()
fig_3d.add_subplot(1,1,1, projection='3d')
plt.scatter(x_values, y_values, z_values, color="purple", linewidth=4)
plt.title("3D image of Scorpius(X, Y, Z)")
plt.show()
````
